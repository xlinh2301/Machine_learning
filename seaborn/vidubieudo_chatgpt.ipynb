```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load bộ dữ liệu "tips"
tips_data = sns.load_dataset("tips")
```

```python
# Scatter Plot (Đồ thị phân tán)
sns.scatterplot(x='total_bill', y='tip', data=tips_data)
plt.title('Scatter plot of Total Bill vs Tip')
plt.show()
```

```python
# Histogram
sns.histplot(tips_data['total_bill'], bins=30, kde=True)
plt.title('Distribution of Total Bill')
plt.show()
```

```python
# Bar Plot (Đồ thị cột)
sns.barplot(x='day', y='total_bill', data=tips_data, ci=None)
plt.title('Average Total Bill by Day')
plt.show()
```

```python
# Box Plot (Đồ thị hộp)
sns.boxplot(x='day', y='total_bill', data=tips_data)
plt.title('Box plot of Total Bill by Day')
plt.show()
```

```python
# Line Plot (Đồ thị đường)
daily_total_bill = tips_data.groupby('day')['total_bill'].mean().reset_index()
sns.lineplot(x='day', y='total_bill', data=daily_total_bill, marker='o')
plt.title('Average Total Bill over Days')
plt.show()
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

# Load the "tips" dataset
tips_data = sns.load_dataset("tips")

# Create a contingency table for two categorical variables (e.g., day and sex)
contingency_table = pd.crosstab(tips_data['day'], tips_data['sex'])

# Perform the Chi-squared test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Print the p-value
print(f'Chi-squared test p-value: {p}')

# Create a heatmap for the contingency table
sns.heatmap(contingency_table, annot=True, cmap='coolwarm', fmt='d')
plt.title('Contingency Table Heatmap')
plt.show()
```

```python
# Pair Plot (Đồ thị cặp)
sns.pairplot(tips_data, hue='sex')
plt.suptitle('Pair Plot of Tips Data', y=1.02)
plt.show()
```
